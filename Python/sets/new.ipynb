### Sets:
Sets are unordered list of unique elements represented using {} brackets

```python
set1={1,4,6,3,5,4,3}
print(set1)
```

```python
set1={1,3,"hello",13.4,1,56,33,33,"hello"}
set1.add("hi")
# set1.clear()
set1.discard(3)
print(set1)
```

```python
#Union

set1={1,3,"hello",13.4,1,56,33,33,"hello"}
set2={1,4,44,90,"yo","no",88.90,3}
print(set1|set2)
print(set1.union(set2))


#Intersection

print(set1&set2)
print(set1.intersection(set2))

#Difference

print(set1-set2)
print(set2.difference(set1))


#Symmetric difference
print(set1^set2)
print(set2.symmetric_difference(set1))

```
