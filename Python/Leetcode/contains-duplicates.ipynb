### 219. Contains Duplicates
Link -> https://leetcode.com/problems/contains-duplicate-ii/

```python
class Solution:
    def containsNearbyDuplicate(self, nums:List[int], k: int) -> bool:
        window = set()
        for i in range(len(nums)):
            if nums[i] in window:
                return True
            
            window.add(nums[i])
            
            if len(window) > k:
                window.remove(nums[i-k])
        
        return False
```
