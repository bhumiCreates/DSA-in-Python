# Leetcode question no#125
 Valid Palidrome : Link -> https://leetcode.com/problems/valid-palindrome/

```python
class Solution:
    def isPalindrome(self, s: str) -> bool:
        rev=""
        result=""
        s=s.lower()  
        l=0
        for i in s:
            if i.isalpha() or i.isdigit():
                l=l+1
                result=result+i
        for j in result:
            rev = j + rev

        if result == rev:
            return True
        else:
            return False    
```
