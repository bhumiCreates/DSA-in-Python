# Leetcode question no#125
 Valid Palidrome : Link -> https://leetcode.com/problems/valid-palindrome/

### With using inbuilt functions

```python
class Solution:
    def isPalindrome(self, s: str) -> bool:
        rev=""
        result=""
        s=s.lower()  
        l=0
        for i in s:
            if i.isalpha() or i.isdigit():
                l=l+1
                result=result+i
        for j in result:
            rev = j + rev

        if result == rev:
            return True
        else:
            return False    
```

### Without using in built functions

```python
class Solution:
    def isAlpha(self,s,i):
        x=ord(s[i])
        if 97<=x<=122 or 48<=x<=57:
            return True
        else:
            return False


    def isPalindrome(self, s: str) -> bool:
        s=s.lower()
        n=len(s)
        i,j=0,n-1

        while i<j:
            if not self.isAlpha(s,j):
                j-=1
                continue
            if not self.isAlpha(s,i):
                i+=1
                continue
            if s[i]==s[j]:
                i+=1
                j-=1

            else:
                return False
            
        return True
```
