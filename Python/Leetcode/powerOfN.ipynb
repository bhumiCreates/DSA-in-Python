### Power of 3
Link -> https://leetcode.com/problems/power-of-three/


```python
class Solution:
    def isPowerOfThree(self, n: int) -> bool:
        if n==0:
            return False
        if n==1:
            return True
        if n==2:
            return False
        if n%3==1:
            return False

        return self.isPowerOfThree(n/3)
```

### Power of four
Link -> https://leetcode.com/problems/power-of-four/

```python
class Solution:
    def isPowerOfFour(self, n: int) -> bool:
        if n==0:      
           return False
        while n%4==0:
           n/=4
        return n==1 
```
