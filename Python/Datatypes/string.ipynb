# String
Stores character , immutable, all the feature of tuple are applicable on strings too.

```python
str="Hello world!"

print(len(str))

print(str)

print(str[2:8])

# print(list(str)) #indirectly mutate string
print(str.lower())

print(str.upper())

print(str.capitalize())

str.replace("world","Bhumi")

print(str[::-1])

```

```python
print(str.find("!"))

print(str.index("world"))

print(str.startswith("Hello"))

print(str.endswith("!"))
```

```python
s="The sky is blue , sun is yellow"

s=s.strip() #trims the space at start and end

s=s.strip("-") #trims the hyphens

print(s.split("e")) # removes the character and splits the parts around it


s=" ".join(s)

# print(s)


```

```python
s="The sky is blue , sun is yellow"

s=s.split()

s.reverse()

s=" ".join(s)

print(s)
```

# Leetcode Problem : 
### #151 Reverse words in a string

Link: https://leetcode.com/problems/reverse-words-in-a-string/


```python
class Solution:
    def reverseWords(self, s: str) -> str:
        s=s.strip().split()[::-1]
        s=" ".join(s)

        return s
```
