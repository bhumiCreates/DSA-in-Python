# Lists :
Variable , dynamic arrays , ordered , mutable , allows duplicates, heterogeneous elements

```python
List1 = [ "Riya", 23,"Rahul",26,"Divya","Yash"] #List
print(List1)
```

```python
print(len(List1))
print(type(List1))
print(List1[5])
print(List1[-1]) #negative indexing
# List1.append(True)
print(List1)
```

```python
l1=["Hello","Riya",23,True,400500.23]
l2=["New York",25,True]

for i in l2 :
    l1.append(i)

print(l1)

#OR

l1.extend(l2) #inbuilt function
print(l1)
```

```python
l1=["Hello","Riya",23,True,400500.23]
l1.insert(2,"New york") #inserts elements at given location 
l1.remove("Hello") #removes value
l1.pop(3) #by default removes last element, removes value of index
l1.clear() #empties list, does not changes the id
l1=[] #changes the id 
print(l1)
print(id(l1))
```

```python
lx=[1,2,4,6,7,3,1,44,2,1,32,4443,99,56,15,64,66,44,]
print(lx.count(44))
print(min(lx))
print(max(lx))
lx.sort() #ascending order
lx.sort(reverse=True) #decending order
print(lx)
```

```python
l1=["Hello","Riya",23,True,400500.23,"Welcome"]
print(l1.index(23))

if "Welcome" in l1:
    print(l1.index("Welcome"))
else:
    print("The element does not exist")
```

### Range

```python
print(list(range(1,18,3))) # syntax: list(range(start,end,difference))
print(list(range(20,6)))
```

### Slicing

```python
ln=[1,2,3,4,5,6,7,8,33,4,2,45,22,23,2,3,22,456,77,556,643]
print(ln[2:10]) 
print(ln[:10]) #by default start index is 0
print(ln[5:]) #by default end index is the index of last element 
print(ln[7:1:-2]) #(start,end,difference)
print(ln[::-1]) #reverse the list
```
