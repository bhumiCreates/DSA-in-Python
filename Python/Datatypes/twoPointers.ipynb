# Two Pointers
Two pointer is a DSA technique where we use two variables (pointers) to traverse a list/array instead of one.

### Check if Array is Palindrome

```python
arr = [1, 2, 3, 2, 1]

left = 0
right = len(arr) - 1

is_palindrome = True

while left < right:
    if arr[left] != arr[right]:
        is_palindrome = False
        break
    left += 1
    right -= 1

print(is_palindrome)
```

### Reverse an Array (Opposite Direction)

```python
arr = [1, 2, 3, 4, 5]

left = 0
right = len(arr) - 1

while left < right:
    arr[left], arr[right] = arr[right], arr[left]
    left += 1
    right -= 1

print(arr)
```
