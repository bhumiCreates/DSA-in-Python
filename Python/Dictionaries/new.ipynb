### Dictionaries
In Dictionaries ,elements are stored in key value pairs where keys are unique but values can be duplicate

```python
dict1={1:"Lakshita",2:"Bhumi"}
print(dict1[1])
dict1[3]="Palak"

del dict1[1]
# dict1.pop(3)
dict1.update({12:"Nice"})
print(dict1)
```

```python
dict2={1:"Lakshita",2:"Bhumi", 3:"Riya", 4:"Rahul"}
for i in dict2.values():
    print(i)

print(list(dict2.values()))
```

```python
dict2={1:"Lakshita",2:"Bhumi", 3:"Riya", 4:"Rahul"}
for i,j in dict2.items():
    print(i,j)

print(dict2.items())
```
