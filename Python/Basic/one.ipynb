```python
# loop will run until the x becomes equal to the i 
x=int(input())
n=int(input())
for i in range (n):
    if i==x:
        break
    print(i,end=" ")
```

# Time complexity:
 Time complexity is a function of time by which the input affects the time taken to execute the program

```python
#time complexity = O(n)
for i in range (n):
    for j in range (x):
        print(i,j,end=" ")
```

```python
#time complexity = O(1)
for i in range(n):
    for j in range(x):
        print(n,x,i,j,)
```

```python
for i in range (n**2): #time complexity = O(n^2)
    print(i,end=" ")

for i in range (n): #time complexity = O(n^2)
    for j in range (n):
        print(i,j,end=" ")

for i in range (n):#time complexity = O(n^x)
    for j in range(x):
        print(i,j,end=" ")

    
```

```python
#time complexity = O(n+x) =linear
for i in range (n):
    print(i,end=" ")
for j in range (x):
    print(j,end=" ")
```

# time complexity O(n^2+n) ~ O(n^2) -> it shows that as we ignore constants in time complexity we also ignore smaller or negligible variable.

```python
n=100 #O(log {base 2} n)
i=1
while i<=n:
    print(i,end=" ")
    i*=2
```

```python
i=100 #O(log{base 3}n )
while i>0:
    print(i,end=" ")
    i//=3
```

# Comparison of time complexities
O(1) < O(log n) < O(n) < O(n log n) < O(n ^ 2) < O(2 ^ n) < O(n!)
